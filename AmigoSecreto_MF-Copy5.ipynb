```python
# Importar Libraries
import pandas as pd
import openpyxl
import copy
from random import choice
```

```python
# Abrir o ficheiro Excel
wb = openpyxl.load_workbook(exc_pessoas)
```

```python
# Abrir a worksheet do ficheiro
ws_pessoas = wb['Pessoas']
```

```python
# Criar uma dataframe com os dados do ficheiro de Excel
# Iniciar uma espécie de apontador. Começa na primeira linha e primeira coluna do ficheiro
data_pessoas = ws_pessoas.values

# Get the first line in file as a header line
columns_pessoas = next(data_pessoas)

# Create a DataFrame based on subsequent lines of data. Use as headers the list obtained in the previous command
df_pessoas = pd.DataFrame(data_pessoas, columns=columns_pessoas)
```

```python
# Sorteio dos amigos secretos
# Opção auto_sorteio = False -> não permite uma pessoa calhar a si própria
# Opção mesmo_grupo = False -> não vai permitir sorteios inter grupos

auto_sorteio = False
mesmo_grupo = False

resortear = True
cont = 0

while resortear == True:
    amigo_secreto = copy.deepcopy(dic_pessoas)
    sorteio = set(amigo_secreto.keys())
    resortear = False
    matches = []
    
    for pessoa in amigo_secreto.keys():
        sorteio_valido = sorteio.copy()

        # Remove do sorteio as pessoas do mesmo grupo
        if not mesmo_grupo:
            num_grupo = amigo_secreto.values()
            grupo = dic_grupos.keys()
            sorteio_valido = {k: amigo_secreto[k] for k in amigo_secreto.keys() - {'Grupo'}}
        elif not auto_sorteio:
            sorteio_valido.discard(pessoa)

        # Retorna a pessoa sorteada (a quem a prenda será oferecida)
        if len(sorteio_valido)>0:
            sorteado = choice(list(sorteio_valido))
            sorteio.discard(sorteado)
        else:
            resortear = True
            cont+=1
        amigo_secreto[pessoa].append(sorteado)
        # O sorteado fica armazenado na 4ª posição da lista correspondente a cada chave do dicionário amigo_secreto
        matches.append([pessoa, sorteado])
        

print(matches)
False
cont - 0
```

```python
# Criar um dataframe com os resultados para posterior escrita em Excel
df_matches = pd.DataFrame(matches, columns=['Gift Giver', 'Gift Receiver'])
```
